```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import sys
import pandas as pd
import xlwt


query_txt = "동서대학교"

f_name = 'C:\\py_temp\\test\\text.txt'
fc_name = 'C:\\py_temp\\test\\text.csv'
fx_name = 'C:\\py_temp\\test\\text.xls'

path = "C:\py_temp\chromedriver\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get("https://www.naver.com")
time.sleep(2)

element = driver.find_element(By.ID,"query")

element.send_keys(query_txt)

driver.find_element(By.ID,'search-btn').click()

driver.find_element(By.LINK_TEXT,'VIEW').click()

driver.find_element(By.LINK_TEXT,'블로그').click()

all_html = driver.page_source 
soup = BeautifulSoup(all_html, 'html.parser')
content_list = soup.find('ul', class_='lst_total')

cnt = 1
num = []
title =[ ]
content=[ ]
date = [ ]
blogNickname=[ ]

orig_stdout = sys.stdout
f = open(f_name , 'a' , encoding='UTF-8')
sys.stdout = f
time.sleep(1)


for i in content_list.find_all('li', 'bx'):
    num.append(cnt)
    print('1.번호:',cnt)
    
    tmp_title = i.find('a', 'api_txt_lines total_tit').get_text()
    title.append(title)
    print('2.제목:',tmp_title.strip())
    
    tmp_content = i.find('div', 'api_txt_lines dsc_txt').get_text()
    content.append(tmp_content)
    print('3.내용:',tmp_content.strip())
    
    tmp_date = i.find('span', 'sub_time sub_txt').get_text()
    date.append(tmp_date)
    print('4.작성날짜:',tmp_date.strip())
    
    tmp_bolgNickname = i.find('a', 'sub_txt sub_name').get_text()
    blogNickname.append(tmp_bolgNickname)
    print('5.블로그 닉네임:',tmp_bolgNickname.strip())
    print("\n")
    
    cnt+=1
    
    if cnt >= 10:
        break

sys.stdout = orig_stdout
f.close()

result = pd.DataFrame()
result['번호'] = num
result['제목'] = title
result['내용'] = content
result['작성날짜'] = date
result['블로그 닉네임'] = blogNickname

result.to_csv(fc_name, encoding="utf-8-sig", index=False)
result.to_excel(fx_name, index=False)
```
